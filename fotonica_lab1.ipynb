```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_function(expression, x, y):
    """Вычисление значения заданной функции"""
    func = sp.lambdify([px, py], expression)
    result = func(x, y)
    return result
```

```python
px, py = sp.symbols("px py")
expr = 20*(px**2 + py**2)**3 - 30*(px**2 + py**2)**2 + 12*(px**2 + py**2) - 1

var = np.linspace(start=-1, stop=1, num=100)
var_left = var - 0.01
var_right = var + 0.01

calc_function_x = calculate_function(expr, var, 0)
calc_function_y = calculate_function(expr, 0, var)

auto_diff_x = expr.diff(px)
auto_diff_y = expr.diff(py)

calc_auto_diff_x = calculate_function(auto_diff_x, var, 0)
calc_auto_diff_y = calculate_function(auto_diff_y, 0, var)

one_side_diff_x = (calculate_function(expr, var, 0) - calculate_function(expr, var_left, 0))*100
one_side_diff_y = (calculate_function(expr, 0, var) - calculate_function(expr, 0, var_left))/0.01
two_side_diff_x = (calculate_function(expr, var_right, 0) - calculate_function(expr, var_left, 0))/0.01
two_side_diff_y = (calculate_function(expr, 0, var_right) - calculate_function(expr, 0, var_left))/0.01

fig1 = plt.figure()
plt.plot(var, calc_function_x, "r", label="По х")
plt.plot(var, calc_function_y, "b", label="По у")
plt.legend(loc="upper right")
plt.title("График функции аберации")


fig2 = plt.figure()
plt.plot(var, calc_auto_diff_x, "r", label="По x")
plt.plot(var, calc_auto_diff_y, "b", label="По y")
plt.legend(loc="upper right")
plt.title("График аналитически вычисленных производных")


fig3 = plt.figure()
plt.plot(var, one_side_diff_x, "r", label="По х")
plt.plot(var, one_side_diff_y, "b", label="По у")
plt.legend(loc="upper right")
plt.title("График односторонней производной")


fig4 = plt.figure()
plt.plot(var, two_side_diff_x, "r", label="По х")
plt.plot(var, two_side_diff_y, "b", label="По у")
plt.legend(loc="upper right")
plt.title("График двусторонней производной")
plt.show()
```
